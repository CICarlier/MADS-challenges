## IMDB Dataset assignment.

```python
import pandas as pd
movies = pd.read_csv('http://bit.ly/imdbratings')
movies.head()
```

### Q1 How many different genre are on the dataset?

```python
nb_genres = len(movies.genre.unique())
nb_genres
```

### Q2 Create a list with the genres

```python
genres = list(movies.genre.unique())
genres
```

### Q3 What is the most popular genre?

```python
count_genres = movies.genre.value_counts()
most_popular_genre = count_genres.idxmax()
most_popular_genre
```

### Q4 How many 'Crime' movies are on the dataframe?

```python
crime_movies = count_genres["Crime"]
crime_movies
```

### Q5 Change the index of the dataframe to "title"

```python
movies.set_index("title", inplace=True)
movies
```

### Q6 Show a list with all the movies of Robert De Niro

```python
RDN = list(movies[movies["actors_list"].str.contains("Robert De Niro")].index)
RDN
```

### Q7 Print a table with all the comedies by Steve Carell. Sort from shorter to longer

```python
Comedies_SC = movies[(movies["genre"] == "Comedy") & (movies["actors_list"].str.contains("Steve Carell"))].sort_values(by="duration")
Comedies_SC
```

### Q8 Print a table with the movies with the highest, and the lowest rating acted by Diane Keaton.

```python
DK = movies[movies["actors_list"].str.contains("Steve Carell")]
max_DK = DK[DK['star_rating']==DK['star_rating'].max()]
min_DK = DK[DK['star_rating']==DK['star_rating'].min()]
DK_max_min = pd.concat([max_DK, min_DK])
DK_max_min
```

### Q9 Make a table with the average time per genre, sorted in descending order.

```python
average_time_genre = movies.groupby("genre")[["duration"]].mean().sort_values(by="duration", ascending=False)
average_time_genre
```

### The RelaxFactor is equal to the value corresponding to each genre in the following dictionary.

```python
RelaxFactor={'Western': 3, 'Adventure': 4, 'Biography': 4, 'Drama': 7, 'Action': 8, 'Crime': 5, 'Mystery': 4, 'Thriller': 5,
 'Fantasy': 6, 'Sci-Fi': 6, 'Comedy' : 10, 'Family': 7, 'Horror': 2, 'Film-Noir': 4, 'Animation': 7, 'History' : 6}
```

```python
movies["RelaxFactor"] = movies.genre.map(RelaxFactor)
movies["Felipe's coefficient"] = (100*(movies.star_rating + movies.RelaxFactor))/movies.duration
columns = ["star_rating", "Felipe's coefficient", "content_rating", "genre", "RelaxFactor", "duration", "actors_list"]
movies = movies[columns].sort_values(by="Felipe's coefficient", ascending=False)
movies
```
